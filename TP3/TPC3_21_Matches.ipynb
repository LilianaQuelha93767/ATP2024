* *Resumo:* Este trabalho tem como objetivo desenvolver um jogo onde há 21 fósforos no início e, a cada jogada, o jogador (utilizador ou computador) pode retirar 1, 2, 3 ou 4 fósforos. Os jogadores jogam alternadamente, e quem retirar o último fósforo perde. O jogo deve oferecer dois modos: no primeiro, o jogador começa e o computador joga em segundo, devendo garantir a vitória; no segundo, o computador começa, e se o jogador cometer um erro, o computador deve assumir a vantagem e vencer.

```python
def jogador_primeiro():
    fosforos = 21
    perdeu = False
    print("\n--- Modo: Jogador primeiro ---")
    print("Existem 21 fósforos. O último a retirar um fósforo perde.")

    while fosforos > 0 and not perdeu:
        # Turno do jogador
        print(f"Fósforos restantes: {fosforos}")
        jogador_move = int(input("Retire entre 1 e 4 fósforos: "))
        
        if jogador_move < 1 or jogador_move > 4:
            print("Escolha incorreta. Por favor, selecione entre 1 e 4 fósforos.")
            continue

        fosforos -= jogador_move

        if fosforos == 0:
            print("Retirou o último fósforo. Perdeste o jogo!")
            perdeu = True
            continue

        # Turno do computador
        computador_move = 5 - jogador_move
        print(f"O computador escolheu retirar: {computador_move}")
        fosforos -= computador_move

        if fosforos == 0:
            print("O computador retirou o último fósforo. O computador perdeu!")
            perdeu = True

def computador_primeiro():
    fosforos = 21
    perdeu = False
    print("\n--- Modo: Computador começa primeiro ---")
    print("Existem 21 fósforos. O último a retirar um fósforo perde.")
    
    while fosforos > 0 and not perdeu:
        # Turno do computador
        if fosforos % 5 != 0:
            computador_move = fosforos % 5
        else:
            computador_move = 1  
        
        print(f"O computador escolheu retirar: {computador_move}")
        fosforos -= computador_move

        if fosforos == 0:
            print("O computador retirou o último fósforo. O computador perdeu!")
            perdeu = True
            continue

        # Turno do jogador
        print(f"Fósforos restantes: {fosforos}")
        jogador_move = int(input("Retire entre 1 e 4 fósforos: "))
        
        # Validação da jogada
        if jogador_move < 1 or jogador_move > 4:
            print("Escolha incorreta. Por favor, selecione entre 1 e 4 fósforos.")
            continue

        fosforos -= jogador_move

        if fosforos == 0:
            print("Retirou o último fósforo. Perdeste o jogo!")
            perdeu = True

def jogo_dos_21_fosforos():
    print("Bem-vindo ao jogo dos 21 fósforos!")
    print("Escolha o modo de jogo:")
    print("1 - Jogador começa primeiro")
    print("2 - Computador começa primeiro")
    
    modo = int(input("Selecione o modo (1 ou 2): "))
    
    if modo == 1:
        jogador_primeiro()
    elif modo == 2:
        computador_primeiro()
    else:
        print("Modo inválido. Por favor, selecione 1 ou 2.")

# Início do jogo
jogo_dos_fosforos()
```
